```python
df = pd.read_csv('housing.csv')
```

```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].mean())
```

```python
lb = LabelEncoder()
df['ocean_proximity'] = lb.fit_transform(df['ocean_proximity'])
```

```python
plt.figure(figsize=(22, 18))
sns.heatmap(df.corr(), annot=True, fmt=".2f", linewidths=0.5, cbar=True)
plt.show()
```

```python
x = df.drop(columns='median_house_value')
y = df['median_house_value']
```

```python
x
```

```python
y
```

```python
y = y.values.reshape(-1,1)
```

```python
s = MinMaxScaler()
x = s.fit_transform(x)
y = s.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test , y_pred)
r2 = r2_score(y_test , y_pred)
```

```python
print('mean_squared_error is '+str(mse))
print('mean_absolute_error is '+str(mae))
print('r2 score is '+str(r2))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5, label='Predicted vs Actual')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Fit')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```
